## MODEL

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, accuracy_score

ev= pd.read_csv('cleaned ev.csv')
```

```python
# Encode 'Battery Type' and 'Charging Mode'
le_battery = LabelEncoder()
ev['battery type encoded'] = le_battery.fit_transform(ev['battery type'])
```

```python
features = ['soc', 'current', 'voltage', 'battery temp', 'ambient temp', 'charging cycles', 'battery type encoded']
```

```python
X1 = ev[features]
y1 = ev['degradation rate']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=100)  # You can tune alpha later
ridge_model.fit(X1_train, y1_train)
```

```python
ridge_predictions = ridge_model.predict(X1_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

r2 = r2_score(y1_test, ridge_predictions)
rmse = np.sqrt(mean_squared_error(y1_test, ridge_predictions))
mae = mean_absolute_error(y1_test, ridge_predictions)

print(f"R² Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")

import pickle

# Save the Ridge model as a pickle file
with open('ridge_model.pkl', 'wb') as file:
    pickle.dump(ridge_model, file)

print("Ridge model saved as 'degradation model_model.pkl'")
```

## MODEL 2

```python

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to 'Charging Mode' and 'Battery Type'
ev['battery type'] = label_encoder.fit_transform(ev['battery type'])

X_duration = ev[['soc', 'voltage', 'current', 'battery temp', 'ambient temp', 'battery type','degradation rate']]
y_duration = ev['charging duration']  # Target variable: Charging Duration

from sklearn.model_selection import train_test_split
X2_train, X2_test, y2_train, y2_test = train_test_split(X_duration, y_duration, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X2_train, y2_train)
y2_pred = lr_model.predict(X2_test)


# Evaluate the model
rmse = mean_squared_error(y2_test, y2_pred, squared=False)  # RMSE
mae = mean_absolute_error(y2_test, y2_pred)  # MAE
r2 = r2_score(y2_test, y2_pred)  # R² Score

# Display the results
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
print(f'R² Score: {r2}')

import pickle

# Save the Linear Regression model as a pickle file
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(lr_model, file)

print("Model saved as 'charging_time_model.pkl'")
```
